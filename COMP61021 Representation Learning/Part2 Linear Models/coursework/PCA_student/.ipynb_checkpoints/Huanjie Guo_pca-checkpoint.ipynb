```python
def print_image(plot, data, shape=(64, 64)):
    plot.imshow(np.flip(data.reshape(shape).T, 1), cmap=plt.get_cmap('gray'))
    plot.axis('off')
```

## Part 1. Implement PCA and Dual PCA algorithms

### <span style="color:red">=========== Assignment 1 ===========</span>

```python
# PCA (with covariance)

def my_pca(X,n_components=None):
    # m is the length of the dataset
    m = X.shape[0]
    
    # centralize X
    X = X.T - np.mean(X,axis=1)    
    
    # get covariance matrix  (X_Format is sample*feature)
    covariance_matrix = 1 / m * np.matmul(X.T, X)
    
 
    # get eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
    
    # rank the eigenvalue and get their idx,  make the pc1 rank in the fist place
    idx = np.argsort(eigenvalues[::-1])
    
    # rank the vector their get top N
    # the final vector is (feature_len*n_components)
    eigenvectors = (eigenvectors[:, idx])[:, :n_components]
    eigenvalue = (eigenvalues[idx])[:n_components]
    # transform
    print(np.matmul(X,eigenvectors))
    return np.matmul(X,eigenvectors),eigenvalue
    
```

### <span style="color:red">=========== End of Assignment 1 ===========</span>

### <span style="color:red">=========== Assignment 2 ===========</span>

```python
# PCA (with SVD)

def my_dual_pca(X,n_component):
    u,s,v = np.linalg.svd(X)
    m = len(u)
    n = len(v)
    recData = np.zeros((m,n))
    
    curSum=0
        
    for k in range(0,n_component):
        uk=u[:,k].reshape(m,1)
        vk=v[k].reshape(1,n)
        recData += s[k]*np.dot(uk,vk)
        curSum += s[k]
    
    return recData,round(curSum/sum(s),4)
```

## Part 3. Image compression

### <span style="color:red">=========== Assignment 4 ===========</span>

```python
X, _ = np.load('./Data/Yale_64x64.npy', allow_pickle=True)

_, num_examples = X.shape
test_sample = 0

test_indices =  np.arange(test_sample, num_examples, 11)
train_indices = np.delete(np.arange(num_examples), test_indices) 

# train_X = 4096*150
train_X = X[:, train_indices]

# test_X = 4096*15
test_X = X[:, test_indices]
```

```python
fig = plt.figure(figsize=(12,30))
grid = plt.GridSpec(20, 8, wspace=0, hspace=0)
# Image compression

compress_ratios = []
compress_thres = 0.75
n_pcs = 1
while True:
    compress_ratios = []
    shouldNext = False
    for i in range(0,len(train_X.T)):
        recData,compress_ratio = my_dual_pca(train_X.T[i].reshape(64,64),n_component=n_pcs)
        if compress_ratio<compress_thres:
            shouldNext = True
            break
        else: compress_ratios.append(compress_ratio)
    if shouldNext==False: break
    n_pcs+=1
print("we use %s components to get a compression ratio higher than %s"%(n_pcs,compress_thres))
for i in range(0,len(train_X.T)):
    recData,compress_ratio = my_dual_pca(train_X.T[i].reshape(64,64),n_component=n_pcs)
    print_image(plt.subplot(grid[i//8,i%8]),recData.reshape((1,64*64)))
```
