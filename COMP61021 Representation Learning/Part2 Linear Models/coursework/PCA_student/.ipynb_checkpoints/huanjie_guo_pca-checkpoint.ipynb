## Part 1. Implement PCA and Dual PCA algorithms

### <span style="color:red">=========== Assignment 1 ===========</span>

````python
# PCA (with covariance)

def my_pca(X):
    # m is the length of the dataset
    m = X.shape[1]
    
    Mean = np.mean(X,axis=1)
    Mean = Mean.reshape(Mean.shape[0],1)
    # centralize X
    X = X-Mean
    
    # get covariance matrix  (X_Format is sample*feature)
    covariance_matrix = 1 / m * np.matmul(X, X.T)
 
    # get eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
    
    # rank the eigenvalue and get their idx,  make the pc1 ranked in the fist place
    idx = np.argsort(eigenvalues[::-1])

    # rank the vector their get top N
    # the final vector is (feature_len*n_components)
    eigenvectors = eigenvectors[:, idx]
    eigenvalue = eigenvalues[idx]
    return eigenvalues,eigenvectors
    
````

### <span style="color:red">=========== End of Assignment 1 ===========</span>

### <span style="color:red">=========== Assignment 2 ===========</span>

````python
# PCA (with SVD)

def my_dual_pca(X):
    # calculate the mean of each feature
    Mean = np.mean(X,axis=1)
    Mean = Mean.reshape(Mean.shape[0],1)
    # centralize X
    X = X-Mean
    
    Y = X.T/np.sqrt(X.shape[1]-1)
    
    # svd
    U,sigma,VT = np.linalg.svd(Y)
    return sigma*sigma,VT.T

````

## Part 2. Visualisation

### <span style="color:red">=========== Assignment 3 ===========</span>

````python
# Import dataset
iris = datasets.load_iris()
X = iris.data.T
y = iris.target
````

````python
# Visualisation
from sklearn.preprocessing import StandardScaler

# use pca to get three pcs
eigenValues,eigenVecs = my_pca(X)

# choose components
n_components = 3

# get projection matrix
Projection_Matrix = eigenVecs[:,:n_components]
print("enigenValues is %s"%(eigenValues[:n_components]))

# Calculate Mean X
Mean = np.mean(X,axis=1)
Mean = Mean.reshape(Mean.shape[0],1)


#Encoding:
z = np.dot(Projection_Matrix.T,X - Mean)

pricipalDf = pd.DataFrame(data=z.T,columns=['PC1','PC2','PC3'])
y = pd.DataFrame(data=y,columns=['target'])
finalDf = pd.concat([pricipalDf,y],axis=1)
# show the pcDF
finalDf
````

````python
fig = plt.figure(figsize=(13,13))
targets = [0,1,2]
colors = ['r','g','b']

# PC1 and PC2
ax1 = fig.add_subplot(2,2,1)
ax1.set_xlabel('PC1',fontsize=10)
ax1.set_ylabel('PC2',fontsize=10)
ax1.set_title('PC1-PC2', fontsize=15)

for target,color in zip(targets,colors):
    indicesToKeep = finalDf['target']==target
    ax1.scatter(finalDf.loc[indicesToKeep,'PC1']
              ,finalDf.loc[indicesToKeep,'PC2']
              ,c=color
              ,s=30)
ax1.legend(targets)

# PC1 and PC3
ax2 = fig.add_subplot(2,2,2)
ax2.set_xlabel('PC1',fontsize=10)
ax2.set_ylabel('PC3',fontsize=10)
ax2.set_title('PC1-PC3', fontsize=15)

for target,color in zip(targets,colors):
    indicesToKeep = finalDf['target']==target
    ax2.scatter(finalDf.loc[indicesToKeep,'PC1']
              ,finalDf.loc[indicesToKeep,'PC3']
              ,c=color
              ,s=30)
ax2.legend(targets)

# PC2 and PC3
ax3 = fig.add_subplot(2,2,3)
ax3.set_xlabel('PC2',fontsize=10)
ax3.set_ylabel('PC3',fontsize=10)
ax3.set_title('PC2-PC3', fontsize=15)

for target,color in zip(targets,colors):
    indicesToKeep = finalDf['target']==target
    ax3.scatter(finalDf.loc[indicesToKeep,'PC2']
              ,finalDf.loc[indicesToKeep,'PC3']
              ,c=color
              ,s=30)
ax3.legend(targets)
````

## Part 3. Image compression

### <span style="color:red">=========== Assignment 4 ===========</span>

````python
X, _ = np.load('./Data/Yale_64x64.npy', allow_pickle=True)

_, num_examples = X.shape
test_sample = 0

test_indices =  np.arange(test_sample, num_examples, 11)
train_indices = np.delete(np.arange(num_examples), test_indices) 

# train_X = 4096*150
train_X = X[:, train_indices]

# test_X = 4096*15
test_X = X[:, test_indices]
````

````python
# Image compression
from sklearn.decomposition import PCA

def getProjectionN(X):
    eigen_value,eigen_vec = my_dual_pca(X)
    n_components = None
    # calculate total variance
    total_variance = sum(eigen_value)

    # accumulation of variance
    cur_variance = 0

    for i in range(0,len(eigen_value)):
        cur_variance += eigen_value[i]
        # when PoV>0.9  let n_components = i+1
        if cur_variance/total_variance>0.9:
            n_components = i+1
            break
    # print("The number of principal conponents for this data set is %s"%(n_components))
    return eigen_vec[:,:n_components]
        
# we get the projection Matrix
Projection_Matrix = getProjectionN(train_X)
# plot the eigen face
fig = plt.figure(figsize=(16,10))
grid = plt.GridSpec(5, 8, wspace=0, hspace=0)
for i in range(0,len(Projection_Matrix.T)):
    print_image(plt.subplot(grid[i//8,i%8]),Projection_Matrix.T[i])
````

````python
# calculate the loss
def image_loss(data, recdata):
    sumError = 0
    D_value = data - recdata
    for i in range(0,len(D_value)):
        sumError += D_value[i]*D_value[i]
    return sumError

losses = []
for i in range(0,len(test_X.T)):
    losses.append(image_loss(test_X.T[i],rec_images.T[i]))
print(losses)
````

## Part 4. Application: face recognition

### <span style="color:red">=========== Assignment 6 ===========</span>

````python
train_X, train_y, test_X, test_y = np.load('./Data/Yale_64x64-part-4-persons.npy', allow_pickle=True)
````

````python
#Comparison
#PCA
from sklearn.decomposition import PCA
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

# get the eigenface of trainX
Projection_Matrix_ = getProjectionN(train_X)

# Calculate Mean train
Mean_train = np.mean(train_X,axis=1)
Mean_train = Mean_train.reshape(Mean_train.shape[0],1)

# Encoding, here is the low-dimensional representations of all the test images
z_train = np.dot(Projection_Matrix_.T,train_X-Mean_train)
n_pcs_Train_X = z_train.T

# Encoding, here is the low-dimensional representations of all the test images
z_test = np.dot(Projection_Matrix_.T,test_X-Mean_train)
n_pcs_Test_X = z_test.T

# svm-pca
clf_pca = svm.SVC(kernel='rbf')
clf_pca.fit(n_pcs_Train_X,train_y)
pred_pca = clf_pca.predict(n_pcs_Test_X)
print("PCA-SVM-pred %s"%pred_pca)
pca_accuracy_score =accuracy_score(y_pred=pred_pca,y_true=test_y)
pca_confusion_matrix = confusion_matrix(y_pred=pred_pca,y_true=test_y)
print("Accuracy score of pca is %.2f"%pca_accuracy_score)
print("confusion matrix of pca : \n",pca_confusion_matrix)

# svm-no-pca
clf_no_pca = svm.SVC(kernel='rbf')
clf_no_pca.fit(train_X.T,train_y)
pred_no_pca = clf_no_pca.predict(test_X.T)
print("No-PCA-SVM-pred %s"%pred_no_pca)
no_pca_accuracy_score = accuracy_score(y_pred=pred_no_pca,y_true=test_y)
no_pca_confusion_matrix = confusion_matrix(y_pred=pred_no_pca,y_true=test_y)
print("Accuracy score of no-pca is %.2f"%no_pca_accuracy_score)
print("confusion matrix of no-pca : \n",no_pca_confusion_matrix)


````
